```python
cols = ["sample_year","sample_month_number"]

df['year'] = df[cols].apply(lambda row: '-'.join(row.values.astype(str)), axis=1)
```

```python
df.year = pd.to_datetime(df.year, format='%Y-%m')
```

```python
df = df.set_index(df.year).sort_index()
```

```python
df = pd.read_csv("Data/df_st.csv")
state_code = list(df.state_code)
combined_df = pd.DataFrame()
for state in state_code:
    temp_df = df[df.state_code == state].T
    temp_df.columns = ['mites']
    temp_df['state'] = state
    temp_df = temp_df.iloc[1:, :]
    combined_df = combined_df.append(temp_df)
```
